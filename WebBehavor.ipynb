```python

spark.sql("DROP TABLE IF EXISTS cx_pilot.airside_heap_pageviews")
pageviews.write.saveAsTable("cx_pilot.airside_heap_pageviews")
```

```python
nasa_pageviews_agg =spark.sql("""
                              select session_id,
                                      --event_id,
                                      --user_id as user_id_org,
                                      min(time) as first_ts, 
                                      max(time) as max_ts, 
                                      count(session_id) as nbr_sessions
                              from cx_pilot.airside_heap_pageviews
                              group by session_id 
                              --,event_id,user_id
                              """)
```

```python
users_by_date = spark.sql("""
                          select
                            Distinct to_date(time) as Session_Date,
                            user_id
                          from
                            cx_pilot.airside_heap_pageviews
                          """)
```

```python
adoption_airside =spark.sql( """
                          select
                            Distinct 
                            pv.User_Id as UserId,
                            to_date(pv.time) as LoginTime,
                          --   pv.landing_page as LogindUrl,
                            'Airside' as Application

                          from
                            cx_pilot.airside_heap_pageviews as pv 
                            """)
```

```python
spark.sql("DROP TABLE IF EXISTS cx_pilot.airside_qualtrics_reviews")
reviews.write.saveAsTable("cx_pilot.airside_qualtrics_reviews")
```

```python
csat_airside =spark.sql( """
                        select
                          distinct
                          QID2_1 as rating,
                          to_date(recordedDate)  as timestamp,
                          'Airside' as appname
                        from
                          cx_pilot.airside_qualtrics_reviews
                          """)
```

```python
spark.sql("DROP TABLE IF EXISTS cx_pilot.csat_airside")
csat_airside.write.saveAsTable("cx_pilot.csat_airside")

spark.sql("DROP TABLE IF EXISTS cx_pilot.adoption_airside")
adoption_airside.write.saveAsTable("cx_pilot.adoption_airside")

spark.sql("DROP TABLE IF EXISTS cx_pilot.airside_pageviews_users_by_date")
users_by_date.write.saveAsTable("cx_pilot.airside_pageviews_users_by_date")

spark.sql("DROP TABLE IF EXISTS cx_pilot.airside_heap_events")
events.write.saveAsTable("cx_pilot.airside_heap_events")

spark.sql("DROP TABLE IF EXISTS cx_pilot.airside_heap_pageviews_agg")
nasa_pageviews_agg.write.saveAsTable("cx_pilot.airside_heap_pageviews_agg")
```
